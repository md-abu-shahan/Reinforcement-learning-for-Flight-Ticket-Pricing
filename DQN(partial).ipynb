```python
rewards = {}
s=[]
price=[]
for i in range(0,12224):
    a=str(file['departure_leg_1_flight_number'][i])
    b=str(file['departure_leg_1_date'][i])
    #c=str(file['departure_leg_1_month'][i])
    d=str(file['departure_leg_1_departure_time'][i])
    e=str(file['departure_leg_1_airline'][i])
    f=str(file['departure_leg_1_from_airport'][i])
    g=str(file['departure_leg_1_to_airport'][i])
    h=str(file['date_time'][i])

    s.append(a+' '+b+' '+d+' '+e+' '+f+' '+g+''+h)
    price.append(file['price'][i])
    rewards[s[i]]=price[i]
    #print(file['Price'][i])
#print(price)
print(rewards) 
#print(rewards['IndiGo22:20BangloreNew Delhi24032019'])
```

```python
def is_terminal_state(current_row_index, current_column_index):
    #print(current_row_index,' ',current_column_index,' ',s)
    if current_row_index + current_column_index not in s:
        return False
    else:
        return True


def get_starting_location():
    current_row_index = np.random.choice(first)
    current_column_index = np.random.choice(second)
    #print(current_row_index, current_column_index)
    return current_row_index, current_column_index

def get_next_location(current_row_index):
    current_column_index = np.random.choice(second)
    #print(current_row_index, current_column_index)
    return current_row_index, current_column_index

def get_next_action(current_row_index, current_column_index, epsilon):
    if np.random.random() < epsilon:
        return np.argmax(q_values[current_row_index, current_column_index])
    else: #choose a random action
        return np.random.randint(2)
```

```python
replay_memory = {}
w = np.random.rand(2)
loss=[]
epsilon = 0.9          
discount_factor = 1
global row_index, column_index
for episode in range(10000):
    row_index, column_index = get_starting_location()
    
    while(True):
        action_index = get_next_action(row_index, column_index, epsilon)
        #print(action_index)
        x=first1[row_index]
        y=second1[column_index]
        
        if not is_terminal_state(x,y):
            row_index, column_index = get_starting_location()
            continue
        if action_index==0:
            reward = -rewards[x+y]
        else:
            reward = float('-inf') 
        
        if replay_memory.get(x+y):
            replay_memory[x+y]=replay_memory[x+y]+1
        else:
            replay_memory[x+y]=1

       #if max(replay_memory.values())==replay_memory[x+y] and replay_memory[x+y]!=1:
          #  break
        
        value=[]
        value.append(reward)

            
        while(True):
            row_index1, column_index1 = get_next_location(row_index)
            action_index1 = get_next_action(row_index1, column_index1, epsilon)
            x1=first1[row_index1]
            y1=second1[column_index1]
            if not is_terminal_state(x1,y1):
                row_index1, column_index1 = get_next_location(row_index)
                continue


            if action_index1==0:
                reward1 = -rewards[x1+y1]
            else:
                reward1 = float('-inf') 


            if replay_memory.get(x1+y1):
                replay_memory[x1+y1]=replay_memory[x1+y1]+1
            else:
                replay_memory[x1+y1]=1

            #if max(replay_memory.values())==replay_memory[x1+y1] and replay_memory[x1+y1]!=1:
                #continue

            value.append(reward1)
            

            while(True):
                row_index2, column_index2 = get_next_location(row_index)
                action_index2 = get_next_action(row_index2, column_index2, epsilon)
                x2=first1[row_index2]
                y2=second1[column_index2]
                if not is_terminal_state(x2,y2):
                    row_index2, column_index2 = get_next_location(row_index)
                    continue


                if action_index2==0:
                    reward2 = -rewards[x2+y2]
                else:
                    reward2 = float('-inf') 
                if replay_memory.get(x2+y2):
                    replay_memory[x2+y2]=replay_memory[x2+y2]+1
                else:
                    replay_memory[x2+y2]=1

                if max(replay_memory.values())==replay_memory[x2+y2] and replay_memory[x2+y2]!=1:
                    continue
                value.append(reward2)

                maxx=max(value)
                
                if maxx == reward:
                    new_q_value = reward + (discount_factor * np.max(q_values[row_index, column_index]))*w 
                    q_values[row_index, column_index, action_index] = new_q_value
                    current_value = reward
                    terget_value = new_q_value
                elif maxx == reward2:
                    new_q_value = reward2 + (discount_factor * np.max(q_values[row_index2, column_index2]))*w
                    q_values[row_index2, column_index2, action_index2] = new_q_value
                    current_value = reward2
                    terget_value = new_q_value
                else:
                    new_q_value = reward1 + (discount_factor * np.max(q_values[row_index1, column_index1]))*w
                    q_values[row_index1, column_index1, action_index1] = new_q_value
                    current_value = reward1
                    terget_value = new_q_value 
                #print(reward1,reward2,reward,current_value)
                #print(terget_value)
                loss.append(math.sqrt(((current_value - terget_value)**2)))
                w = w * [1, terget_value]
                if action_index2==0:
                    break;
            if action_index1==0:
                break
        if action_index==0:
            break;
    print(episode)
q_values
```

```python
def analysis(flight_no,date,departure_time,airline,source,destination):
    unique_time=[]
    opservation=[]
    for i in range(0,12224):
        if(file['departure_leg_1_airline'][i]==airline and str(file['departure_leg_1_date'][i])==date
           and str(file['departure_leg_1_flight_number'][i])==flight_no and file['departure_leg_1_departure_time'][i]==departure_time
           and file['departure_leg_1_from_airport'][i]==source and file['departure_leg_1_to_airport'][i]==destination):
            unique_time.append(file['date_time'][i])
            opservation.append(file['observation_count'][i])
    
    cols=[]
    string=flight_no+' '+date+' '+departure_time+' '+airline+' '+source+' '+destination;
    for i in range(len(unique_time)):
        for j in range(len(second)):
            #print(s)
            if (second1[j] == unique_time[i]):
                cols.append(j)
                break
    
    for i in range(len(first)):
        if first1[i] == string:
            row=i
            break
    
    val=[]
    for i in range(len(cols)):
        val.append(q_values[row][cols[i]][0])
    
    
    min_price=max(val)
    result=[]
    for i in range(len(cols)):
        #print(departure_time,' ',airline,' ',unique_time[i],' ',opservation[i],' ',source,' ',destination,' ',date,' : ',end=' ')
        old_month=int(unique_time[i][5]+unique_time[i][6])
        old_date=int(unique_time[i][8]+unique_time[i][9])
        old_time=int(unique_time[i][11]+unique_time[i][12])
        old_year=int(unique_time[i][2]+unique_time[i][3])
        current_year=int(3)
        current_month=int(1)
        current_date=int(date)
        if(departure_time[1]==':'):
            current_time=int(departure_time[0])
        else:
            current_time=int(departure_time[0]+departure_time[1])
        if(min_price==val[i]):
            result.append(((((((current_year-old_year)*12)-old_month+current_month)*30)-old_date+current_date)*24)+(current_time-old_time))
            #print('b')
    return result,unique_time
```

```python
def output(flight_no,date,departure_time,airline,source,destination,search_time,search_date,results):
    old_month=int(1)
    old_date=int(search_date)
    old_time=int(search_time)
    old_year=int(3)
    current_year=int(3)
    current_month=int(1)
    current_date=int(date)
    if(departure_time[1]==':'):
        current_time=int(departure_time[0])
    else:
        current_time=int(departure_time[0]+departure_time[1])
    result=((((((current_year-old_year)*12)-old_month+current_month)*30)-old_date+current_date)*24)+(current_time-old_time)
    if(result in results):
        return 'Buy'
    else:
        return 'Wait'
```

```python
flight_no='192'
departure_date='3'
departure_time='9:45'
airline='American Airlines'
source='LAX'
destination='BOS'
search_date='1'
search_time='12'

departure_time=departure_time+'p'
string=flight_no+' '+departure_date+' '+departure_time+' '+airline+' '+source+' '+destination;

#print(string)
answer,unique_time=analysis(flight_no,departure_date,departure_time,airline,source,destination)
#print(unique_time)
out=output(flight_no,departure_date,departure_time,airline,source,destination,search_time,search_date,answer)
print(out)
```
